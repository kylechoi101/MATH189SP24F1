# Renaming

```python
import os
import re

# List of California counties in alphabetical order
counties = [
    'Alameda', 'Alpine', 'Amador', 'Butte', 'Calaveras', 'Colusa', 'Contra Costa', 
    'Del Norte', 'El Dorado', 'Fresno', 'Glenn', 'Humboldt', 'Imperial', 'Inyo', 
    'Kern', 'Kings', 'Lake', 'Lassen', 'Los Angeles', 'Madera', 'Marin', 'Mariposa', 
    'Mendocino', 'Merced', 'Modoc', 'Mono', 'Monterey', 'Napa', 'Nevada', 'Orange', 
    'Placer', 'Plumas', 'Riverside', 'Sacramento', 'San Benito', 'San Bernardino', 
    'San Diego', 'San Francisco', 'San Joaquin', 'San Luis Obispo', 'San Mateo', 
    'Santa Barbara', 'Santa Clara', 'Santa Cruz', 'Shasta', 'Sierra', 'Siskiyou', 
    'Solano', 'Sonoma', 'Stanislaus', 'Sutter', 'Tehama', 'Trinity', 'Tulare', 
    'Tuolumne', 'Ventura', 'Yolo', 'Yuba'
]

# Directory containing the files
directory = r'C:\Users\khali\Desktop\Math 189 Project Folder'  # Change this to the path where your files are located

# Get the list of files in the directory
files = os.listdir(directory)

# Filter the files that match the naming pattern
files_to_rename = [f for f in files if f.startswith("State of California Department of Justice - OpenJustice (")]

# Custom sorting function to handle the numeric and special character order
def custom_sort_key(file_name):
    match = re.match(r'.*\((\d+)([a-z]?)\)', file_name) #This was need as we missed a file and had to go back and rename 4a
    if match:
        num = int(match.group(1))
        char = match.group(2)
        return (num, char)
    return (float('inf'), '')

# Sort the files using the custom key
files_to_rename.sort(key=custom_sort_key)

# Check if the number of files matches the number of counties
if len(files_to_rename) != len(counties):
    print("Error: The number of files does not match the number of counties.")
else:
    # Rename the files
    for i, file_name in enumerate(files_to_rename):
        old_path = os.path.join(directory, file_name)
        new_file_name = counties[i] + os.path.splitext(file_name)[1]  # Preserve the file extension
        new_path = os.path.join(directory, new_file_name)
        os.rename(old_path, new_path)
        print(f'Renamed: {old_path} -> {new_path}')
```

# Creating 2 CSV Files

```python
import os
import pandas as pd
import numpy as np

# Directory containing the files
directory = r'C:\Users\khali\Desktop\Math 189 Project Folder'  # Change this to the path where your files are located


data_2011 = {'County': []}
data_2017 = {'County': []}


files = os.listdir(directory)

# Set of all crime names
all_crimes = set()

#Collect all crime names
for file_name in files:
    if file_name.endswith('.csv'):
        file_path = os.path.join(directory, file_name)
        df = pd.read_csv(file_path)
        crime_names = df.iloc[:, 0].tolist()  # Column A has the crime names
        all_crimes.update(crime_names)

#Collect statistics
for file_name in files:
    if file_name.endswith('.csv'):
        county_name = os.path.splitext(file_name)[0]  # Get county name from file name
        file_path = os.path.join(directory, file_name)
        
        
        df = pd.read_csv(file_path)
        
        # Extract the crime names and statistics
        crime_names = df.iloc[:, 0].tolist()  # Column A has the crime names
        stats_2011 = df.iloc[:, 1].tolist()   # Column B has the 2011 statistics
        stats_2017 = df.iloc[:, 2].tolist()   # Column C has the 2017 statistics
        
        # Fill in the data for the current county
        data_2011['County'].append(county_name)
        data_2017['County'].append(county_name)
        
        # Create a dictionary for the current file's data
        crime_data_2011 = dict(zip(crime_names, stats_2011))
        crime_data_2017 = dict(zip(crime_names, stats_2017))
        
        # Add data to the main dictionary
        for crime in all_crimes:
            if crime not in data_2011:
                data_2011[crime] = []
            if crime not in data_2017:
                data_2017[crime] = []
            
            data_2011[crime].append(crime_data_2011.get(crime, np.nan))
            data_2017[crime].append(crime_data_2017.get(crime, np.nan))


df_2011 = pd.DataFrame(data_2011)
df_2017 = pd.DataFrame(data_2017)


df_2011.to_csv('combined_2011.csv', index=False)
df_2017.to_csv('combined_2017.csv', index=False)
```

```python
import pandas as pd

# File paths for the combined CSV files
file_path_2011 = r'C:\Users\khali\Desktop\Math 189 Project Folder\combined_2011.csv'
file_path_2017 = r'C:\Users\khali\Desktop\Math 189 Project Folder\combined_2017.csv'

# Load the combined CSV files
df_2011 = pd.read_csv(file_path_2011)
df_2017 = pd.read_csv(file_path_2017)

# Define a function to clean the DataFrame
def clean_dataframe(df):
    # Remove rows where the first column (County name) is missing
    df_cleaned = df.dropna(subset=[df.columns[0]])
    # Remove all empty columns
    df_cleaned = df_cleaned.dropna(axis=1, how='all')
    return df_cleaned


df_2011_cleaned = clean_dataframe(df_2011)
df_2017_cleaned = clean_dataframe(df_2017)


cleaned_file_path_2011 = r'C:\Users\khali\Desktop\Math 189 Project Folder\cleaned_combined_2011.csv'
cleaned_file_path_2017 = r'C:\Users\khali\Desktop\Math 189 Project Folder\cleaned_combined_2017.csv'

df_2011_cleaned.to_csv(cleaned_file_path_2011, index=False)
df_2017_cleaned.to_csv(cleaned_file_path_2017, index=False)


df_2011_cleaned.head(), df_2017_cleaned.head()
```
