# Cleaning graduation data

#### 2011 Cohort Data

```python
import pandas as pd

# Step 1: Read the .txt file
file_path = '2011drop.txt'  # Path to your .txt file
df = pd.read_csv(file_path, delimiter='\t')  # Adjust delimiter as needed

# Step 2: Process the data (if needed)
df.columns = [col.strip() for col in df.columns]  # Example: Strip whitespace from column names
df.dropna(inplace=True)  # Example: Drop rows with missing values

# Step 3: Save the DataFrame as an Excel file
csv_file_path = '2011drop.csv'
df.to_csv(csv_file_path, index=False)

# Step 4: Load the Excel file into a new DataFrame
df_csv = pd.read_csv(csv_file_path)

# Display the DataFrame
df_csv.head()
```

# Cleaning Crime Data

#### 2011 Violent Crime Data

```python
# Load the Excel file into a DataFrame
file_path = 'Table8_2011.xls'
df = pd.read_excel(file_path, header=None, skiprows=4, nrows=461, usecols="A:L")

# Set the column names based on the row A5 to L5
column_names = df.iloc[0].str.replace('\n', ' ')
df.columns = column_names

# Drop the first row since it contains the column names
df = df.drop(0)

# Reset index
df.reset_index(drop=True, inplace=True)

#Save the DataFrame to a CSV file
csv_file_path = 'Table8_2011.csv'
df.to_csv(csv_file_path, index=False)

# Display the DataFrame
df.head()
```
